## Linear Regression on Simple Dataset

This graph visualizes the relationship between TV advertising spend (in â‚¹000s) and sales (also in â‚¹000s) using simple linear regression.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Load dataset
data = {
    'TV': [50, 80, 100, 120, 150, 170, 200, 220, 250, 270],
    'Sales': [7, 9, 12, 15, 18, 21, 25, 26, 30, 32]
}
df = pd.DataFrame(data)

# Step 2: Prepare data
X = df[['TV']]  # Feature must be 2D
y = df['Sales']

# Step 3: Train model
model = LinearRegression()
model.fit(X, y)

# Step 4: Print model details
print("Coefficient (Slope):", model.coef_[0])
print("Intercept:", model.intercept_)

# Step 5: Predict and visualize
y_pred = model.predict(X)

plt.scatter(X, y, color='blue', label="Actual")
plt.plot(X, y_pred, color='red', label="Regression Line")
plt.xlabel("TV Advertising Spend (â‚¹k)")
plt.ylabel("Sales (â‚¹k)")
plt.title("Simple Linear Regression: TV vs Sales")
plt.legend()
plt.grid(True)
plt.show()
```

**ðŸ“Š Insight Summary**
The graph shows a positive linear relationship between TV ad spend and sales. As companies spend more on TV advertising, sales generally increase.

The red line is the regression line, showing predicted sales.

The blue dots are actual sales, and they mostly align well with the line â€” indicating a good model fit.

Small deviations suggest other factors may also affect sales.

**Conclusion:**
TV advertising has a clear, positive impact on sales, based on this simple linear regression model.
